```python
# arithmetic operations on Series :

from pandas import Series

series1 = Series([1, 2, 3, 4, 5])
series2 = Series([100, 200, 300, 400, 500, 600])
```

```python
# Addition :

series1 + series2

# if an index of series doesnot have a match in another series then, 
# it display NaN(Not a Number).
```

```python
# arithmetic operations on DataFrames :

from pandas import DataFrame
data1 = {'speed': [100, 200, 300],
        'temp': [34, 23, 56],}
frame1 = DataFrame(data1)
frame1
```

```python
data2 = {'speed': [400, 500, 500],
        'temp': [84, 53, 76],
        'humidity': [12, 45, 78]
        }
frame2 = DataFrame(data2)
frame2
```

```python
# Addition

frame1 + frame2

# if an index of DataFrame doesnot have a match in another DataFrame then, 
# it display NaN(Not a Number).
```
