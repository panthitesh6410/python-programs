```python
# reindexing :
# to create new object to data

# reindexing a series :

from pandas import Series
obj = Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
obj
```

```python
obj = obj.reindex(['i', 'j', 'k', 'l', 'm'])   # reindexing
obj
```

```python
# reindexing a dataframe :

from pandas import DataFrame
data = {'name':['john', 'tim', 'rob'],
       'age': [12, 13, 14],
       'salary': [100, 200, 300]
       }
frame = DataFrame(data)
frame
```

```python
# reindexing frame (row):
frame = frame.reindex([2, 1, 0])
frame
```

```python
# reindexing frame (col):
fields = ['age', 'salary', 'name']
frame = frame.reindex(columns=fields)
```
