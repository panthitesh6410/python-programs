```python
# dataframes is a datastr. used in pandas
# it has two indices (row ans col)
# it has a tabular structure with rows and columns

# creating a dataframe :
# first we create a dictionary then make dataframe of it

data = {'name': ['john', 'tim', 'rob'], 
        'age': [19, 20, 21],
        'salary': [1000, 2000, 3000]
       }
```

```python
# to convert this data into dataframe :

from pandas import DataFrame

# creating frame object
frame = DataFrame(data)
```

```python
frame         # displaying dataframe.
```

```python
# changing column sequence in DataFrame:

new_frame = DataFrame(data, columns=['name', 'salary', 'age', 'address'])
```

```python
new_frame
```

```python
# how to retrieve data from dataframe in form of series :

new_frame['salary']
```

```python
# to retrieve a row using it's index:

new_frame.iloc[2]   
```

```python
# modifying a dataframe :
# assigning values to an attribute

new_frame['address'] = 'jabalpur'
# above will change the address field of all rows as jabalour
```

```python
new_frame
```

```python
 # to add a particular field in a dataframe :

new_frame['degree'] = 'Btech'    
```

```python
new_frame
```

```python
# how to transpose the dataframe :

new_frame.T
```
