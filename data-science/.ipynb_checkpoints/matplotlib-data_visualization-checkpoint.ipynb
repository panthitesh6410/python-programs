```python
import pandas as pd    # used to import data from csv files
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title('squares plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title('squares plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.legend(['this is y', 'this is z'])
plt.show()
```

```python
# Analysis :
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# compare population growth in US and China :
# isolating us and china data seperately from csv_file(dataframe) :

us = data[data.country == 'United States']
us
```

```python
china = data[data.country == "China"]
china
```

```python
# plotting US and China population :

plt.plot(us.year, us.population/10**6)     # to get population in million(s)
plt.plot(china.year, china.population/10**6)     # to get population in million(s)
plt.xlabel('year')
plt.ylabel('population(mn)')
plt.legend(['US Population', 'China Population'])
plt.plot()
```

```python
us.population
```

```python
us.population / us.population.iloc[0] * 100
```

```python
# to find percentage(%) growth :

plt.plot(us.year, us.population / us.population.iloc[0] * 100)     # to get population in million(s)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)     # to get population in million(s)
plt.xlabel('year')
plt.ylabel('population(mn)')
plt.legend(['US Population', 'China Population growth %'])
plt.plot()
```
