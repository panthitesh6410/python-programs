```python
# converting dictionary to series :

salary = {'john': 2000, 'bob': 4000, 'tim': 3000}         # a dictionary
```

```python
salary
```

```python
from pandas import Series
se_salary = Series(salary)    # dictionary into series 
```

```python
se_salary
```

```python
se_salary['bob']     # accessing item using index (here key becomes index)
```
