```python
from pandas import DataFrame
data = {'name':['john', 'tim', 'rob'],
       'age': [12, 13, 14],
       'salary': [100, 200, 300]
       }
frame = DataFrame(data)
frame
```

```python
# to delete / drop rows :

frame1 = frame.drop([2, 0])
frame1
```

```python
# to delete / drop columns :

frame2 = frame.drop('age', axis=1)     # y axis (i.e. column)
frame2
```
